```python
from pymongo import MongoClient
client = MongoClient("mongodb://root:mongo@mongo_service:27017/")
db = client["Universidade"]
collection = db["Estudantes"]
```

```python
estudantes = [
    {"nome": "Ana", "idade": 22, "curso": "Engenharia", "email": "ana@email.com"},
    {"nome": "Bruno", "idade": 24, "curso": "Direito", "email": "bruno@email.com"},
    {"nome": "Carla", "idade": 21, "curso": "Engenharia", "email": "carla@email.com"},
    {"nome": "Daniel", "idade": 23, "curso": "Medicina", "email": "daniel@email.com"},
    {"nome": "Eduarda", "idade": 25, "curso": "Direito", "email": "eduarda@email.com"}
]
collection.insert_many(estudantes)
```

```python
resultado = collection.aggregate([
    {"$group": {"_id": "$curso", "total": {"$sum": 1}}}
])
for r in resultado:
    print(r)
```

```python
collection.update_one({"nome": "Ana"}, {"$set": {"curso": "Arquitetura"}})
```

```python
collection.delete_one({"nome": "Daniel"})
```

```python

for estudante in collection.find():
    print(estudante)
```
